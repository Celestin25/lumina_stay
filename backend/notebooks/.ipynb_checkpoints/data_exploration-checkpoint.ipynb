# LuminaStay: Data Exploration & Analysis
## Comprehensive Analysis of Morocco Housing Dataset

This notebook explores the dataset to understand patterns, distributions, and relationships in Moroccan housing data.

```python
# Cell 1: Imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

# Set style
sns.set_style("whitegrid")
sns.set_palette("husl")
plt.rcParams['figure.figsize'] = (14, 7)
plt.rcParams['font.size'] = 11

print("‚úÖ Libraries imported successfully!")
```

```python
# Cell 2: Load Dataset
df = pd.read_csv("../morocco_housing.csv")

print("üìä Dataset Overview:")
print(f"   Shape: {df.shape[0]} samples √ó {df.shape[1]} features")
print(f"\nüîç First 5 rows:")
display(df.head())

print(f"\nüìã Column Names:")
print(df.columns.tolist())
```

```python
# Cell 3: Dataset Information
print("‚ÑπÔ∏è  Dataset Information:\n")
df.info()

print("\nüîç Missing Values:")
missing = df.isnull().sum()
if missing.sum() == 0:
    print("   ‚úÖ No missing values detected!")
else:
    print(missing[missing > 0])
```

```python
# Cell 4: Statistical Summary
print("üìà Statistical Summary:\n")
display(df.describe())

print("\nüìä Categorical Features Summary:")
categorical_cols = ['City', 'Neighborhood', 'Property_Type']
for col in categorical_cols:
    print(f"\n{col}: {df[col].nunique()} unique values")
    print(df[col].value_counts().head())
```

```python
# Cell 5: Price Distribution Analysis
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

# Histogram
axes[0].hist(df['Price_MAD'], bins=50, edgecolor='black', alpha=0.7, color='steelblue')
axes[0].axvline(df['Price_MAD'].mean(), color='r', linestyle='--', lw=2, label=f'Mean: {df["Price_MAD"].mean():,.0f}')
axes[0].axvline(df['Price_MAD'].median(), color='g', linestyle='--', lw=2, label=f'Median: {df["Price_MAD"].median():,.0f}')
axes[0].set_xlabel('Price (MAD)', fontsize=12, fontweight='bold')
axes[0].set_ylabel('Frequency', fontsize=12, fontweight='bold')
axes[0].set_title('Price Distribution', fontsize=13, fontweight='bold')
axes[0].legend()
axes[0].grid(True, alpha=0.3)

# Box plot
axes[1].boxplot(df['Price_MAD'], vert=True)
axes[1].set_ylabel('Price (MAD)', fontsize=12, fontweight='bold')
axes[1].set_title('Price Box Plot', fontsize=13, fontweight='bold')
axes[1].grid(True, alpha=0.3, axis='y')

# Violin plot
axes[2].violinplot(df['Price_MAD'], vert=True, showmeans=True, showmedians=True)
axes[2].set_ylabel('Price (MAD)', fontsize=12, fontweight='bold')
axes[2].set_title('Price Violin Plot', fontsize=13, fontweight='bold')
axes[2].grid(True, alpha=0.3, axis='y')

plt.tight_layout()
plt.savefig('price_distribution.png', dpi=300, bbox_inches='tight')
plt.show()

print("‚úÖ Price distribution plots saved!")
print(f"\nüí∞ Price Statistics:")
print(f"   Mean: {df['Price_MAD'].mean():,.0f} MAD")
print(f"   Median: {df['Price_MAD'].median():,.0f} MAD")
print(f"   Std Dev: {df['Price_MAD'].std():,.0f} MAD")
print(f"   Min: {df['Price_MAD'].min():,.0f} MAD")
print(f"   Max: {df['Price_MAD'].max():,.0f} MAD")
```

```python
# Cell 6: Price by City
plt.figure(figsize=(14, 7))
sns.boxplot(data=df, x='City', y='Price_MAD', palette='Set2')
plt.xlabel('City', fontsize=13, fontweight='bold')
plt.ylabel('Price (MAD)', fontsize=13, fontweight='bold')
plt.title('Price Distribution by City\n(Which city has the highest property prices?)', 
          fontsize=15, fontweight='bold', pad=15)
plt.xticks(rotation=45)
plt.grid(axis='y', alpha=0.3, linestyle='--')
plt.tight_layout()
plt.savefig('price_by_city.png', dpi=300, bbox_inches='tight')
plt.show()

print("‚úÖ Price by city plot saved!")
print(f"\nüìç Average Price by City:")
city_stats = df.groupby('City')['Price_MAD'].agg(['mean', 'median', 'count']).round(0)
city_stats = city_stats.sort_values('mean', ascending=False)
print(city_stats)
```

```python
# Cell 7: Price by Property Type
plt.figure(figsize=(14, 7))
sns.violinplot(data=df, x='Property_Type', y='Price_MAD', palette='Set3')
plt.xlabel('Property Type', fontsize=13, fontweight='bold')
plt.ylabel('Price (MAD)', fontsize=13, fontweight='bold')
plt.title('Price Distribution by Property Type\n(How do prices vary across property types?)', 
          fontsize=15, fontweight='bold', pad=15)
plt.xticks(rotation=45)
plt.grid(axis='y', alpha=0.3, linestyle='--')
plt.tight_layout()
plt.savefig('price_by_property_type.png', dpi=300, bbox_inches='tight')
plt.show()

print("‚úÖ Price by property type plot saved!")
print(f"\nüè† Average Price by Property Type:")
type_stats = df.groupby('Property_Type')['Price_MAD'].agg(['mean', 'median', 'count']).round(0)
type_stats = type_stats.sort_values('mean', ascending=False)
print(type_stats)
```

```python
# Cell 8: Correlation Heatmap
numerical_cols = df.select_dtypes(include=[np.number]).columns
correlation_matrix = df[numerical_cols].corr()

plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, fmt='.2f', cmap='coolwarm', 
            square=True, linewidths=0.5, cbar_kws={"shrink": 0.8})
plt.title('Feature Correlation Heatmap\n(How are features related to each other?)', 
          fontsize=15, fontweight='bold', pad=15)
plt.tight_layout()
plt.savefig('correlation_heatmap.png', dpi=300, bbox_inches='tight')
plt.show()

print("‚úÖ Correlation heatmap saved!")
print(f"\nüîó Top Correlations with Price:")
price_corr = correlation_matrix['Price_MAD'].sort_values(ascending=False)
print(price_corr[price_corr.index != 'Price_MAD'])
```

```python
# Cell 9: Size vs Price Analysis
plt.figure(figsize=(14, 7))
scatter = plt.scatter(df['Size_m2'], df['Price_MAD'], 
                     c=df['Price_MAD'], s=50, alpha=0.6, 
                     cmap='viridis', edgecolors='black', linewidth=0.5)
plt.colorbar(scatter, label='Price (MAD)')
plt.xlabel('Property Size (m¬≤)', fontsize=13, fontweight='bold')
plt.ylabel('Price (MAD)', fontsize=13, fontweight='bold')
plt.title('Property Size vs Price\n(Larger properties command higher prices)', 
          fontsize=15, fontweight='bold', pad=15)
plt.grid(True, alpha=0.3, linestyle='--')

# Add trend line
z = np.polyfit(df['Size_m2'], df['Price_MAD'], 1)
p = np.poly1d(z)
plt.plot(df['Size_m2'], p(df['Size_m2']), "r--", lw=2, label='Trend Line')
plt.legend()

plt.tight_layout()
plt.savefig('size_vs_price.png', dpi=300, bbox_inches='tight')
plt.show()

print("‚úÖ Size vs price plot saved!")
print(f"\nüìè Size-Price Correlation: {df['Size_m2'].corr(df['Price_MAD']):.4f}")
```

```python
# Cell 10: Feature Distributions
fig, axes = plt.subplots(2, 3, figsize=(18, 12))
fig.suptitle('Feature Distributions', fontsize=16, fontweight='bold', y=1.00)

# Bedrooms
axes[0, 0].hist(df['Bedrooms'], bins=range(1, df['Bedrooms'].max()+2), 
                edgecolor='black', alpha=0.7, color='skyblue')
axes[0, 0].set_xlabel('Number of Bedrooms', fontweight='bold')
axes[0, 0].set_ylabel('Frequency', fontweight='bold')
axes[0, 0].set_title('Bedroom Distribution', fontweight='bold')
axes[0, 0].grid(True, alpha=0.3)

# Bathrooms
axes[0, 1].hist(df['Bathrooms'], bins=range(1, df['Bathrooms'].max()+2), 
                edgecolor='black', alpha=0.7, color='coral')
axes[0, 1].set_xlabel('Number of Bathrooms', fontweight='bold')
axes[0, 1].set_ylabel('Frequency', fontweight='bold')
axes[0, 1].set_title('Bathroom Distribution', fontweight='bold')
axes[0, 1].grid(True, alpha=0.3)

# Size
axes[0, 2].hist(df['Size_m2'], bins=50, edgecolor='black', alpha=0.7, color='green')
axes[0, 2].set_xlabel('Size (m¬≤)', fontweight='bold')
axes[0, 2].set_ylabel('Frequency', fontweight='bold')
axes[0, 2].set_title('Size Distribution', fontweight='bold')
axes[0, 2].grid(True, alpha=0.3)

# City
df['City'].value_counts().plot(kind='bar', ax=axes[1, 0], 
                               edgecolor='black', alpha=0.7, color='purple')
axes[1, 0].set_xlabel('City', fontweight='bold')
axes[1, 0].set_ylabel('Count', fontweight='bold')
axes[1, 0].set_title('Properties by City', fontweight='bold')
axes[1, 0].tick_params(axis='x', rotation=45)
axes[1, 0].grid(True, alpha=0.3, axis='y')

# Property Type
df['Property_Type'].value_counts().plot(kind='bar', ax=axes[1, 1], 
                                        edgecolor='black', alpha=0.7, color='orange')
axes[1, 1].set_xlabel('Property Type', fontweight='bold')
axes[1, 1].set_ylabel('Count', fontweight='bold')
axes[1, 1].set_title('Properties by Type', fontweight='bold')
axes[1, 1].tick_params(axis='x', rotation=45)
axes[1, 1].grid(True, alpha=0.3, axis='y')

# Amenities
amenities = ['Has_Pool', 'Has_Garden', 'Is_Furnished']
amenity_counts = df[amenities].sum()
axes[1, 2].bar(amenities, amenity_counts, edgecolor='black', alpha=0.7, color='teal')
axes[1, 2].set_xlabel('Amenity', fontweight='bold')
axes[1, 2].set_ylabel('Count', fontweight='bold')
axes[1, 2].set_title('Amenity Distribution', fontweight='bold')
axes[1, 2].tick_params(axis='x', rotation=45)
axes[1, 2].grid(True, alpha=0.3, axis='y')

plt.tight_layout()
plt.savefig('feature_distributions.png', dpi=300, bbox_inches='tight')
plt.show()

print("‚úÖ Feature distribution plots saved!")
```

```python
# Cell 11: Geographic Distribution
plt.figure(figsize=(14, 10))
scatter = plt.scatter(df['Longitude'], df['Latitude'], 
                     c=df['Price_MAD'], s=80, alpha=0.6, 
                     cmap='plasma', edgecolors='black', linewidth=0.5)
cbar = plt.colorbar(scatter, label='Price (MAD)')
cbar.ax.tick_params(labelsize=11)

plt.xlabel('Longitude', fontsize=13, fontweight='bold')
plt.ylabel('Latitude', fontsize=13, fontweight='bold')
plt.title('Geographic Distribution of Properties\n(Property locations colored by price)', 
          fontsize=15, fontweight='bold', pad=15)
plt.grid(True, alpha=0.3, linestyle='--')

# Add city labels
city_centers = df.groupby('City')[['Latitude', 'Longitude']].mean()
for city, coords in city_centers.iterrows():
    plt.annotate(city, (coords['Longitude'], coords['Latitude']), 
                fontsize=12, fontweight='bold', 
                bbox=dict(boxstyle='round,pad=0.5', facecolor='yellow', alpha=0.7))

plt.tight_layout()
plt.savefig('geographic_distribution.png', dpi=300, bbox_inches='tight')
plt.show()

print("‚úÖ Geographic distribution plot saved!")
```

```python
# Cell 12: Amenities Impact on Price
fig, axes = plt.subplots(1, 3, figsize=(18, 6))

# Pool impact
pool_data = df.groupby('Has_Pool')['Price_MAD'].mean()
axes[0].bar(['No Pool', 'Has Pool'], pool_data.values, 
            edgecolor='black', alpha=0.7, color=['lightcoral', 'lightgreen'])
axes[0].set_ylabel('Average Price (MAD)', fontweight='bold')
axes[0].set_title('Impact of Pool on Price', fontweight='bold')
axes[0].grid(axis='y', alpha=0.3)
for i, v in enumerate(pool_data.values):
    axes[0].text(i, v + 50000, f'{v:,.0f}', ha='center', fontweight='bold')

# Garden impact
garden_data = df.groupby('Has_Garden')['Price_MAD'].mean()
axes[1].bar(['No Garden', 'Has Garden'], garden_data.values, 
            edgecolor='black', alpha=0.7, color=['lightcoral', 'lightgreen'])
axes[1].set_ylabel('Average Price (MAD)', fontweight='bold')
axes[1].set_title('Impact of Garden on Price', fontweight='bold')
axes[1].grid(axis='y', alpha=0.3)
for i, v in enumerate(garden_data.values):
    axes[1].text(i, v + 50000, f'{v:,.0f}', ha='center', fontweight='bold')

# Furnished impact
furnished_data = df.groupby('Is_Furnished')['Price_MAD'].mean()
axes[2].bar(['Unfurnished', 'Furnished'], furnished_data.values, 
            edgecolor='black', alpha=0.7, color=['lightcoral', 'lightgreen'])
axes[2].set_ylabel('Average Price (MAD)', fontweight='bold')
axes[2].set_title('Impact of Furnishing on Price', fontweight='bold')
axes[2].grid(axis='y', alpha=0.3)
for i, v in enumerate(furnished_data.values):
    axes[2].text(i, v + 50000, f'{v:,.0f}', ha='center', fontweight='bold')

plt.tight_layout()
plt.savefig('amenities_impact.png', dpi=300, bbox_inches='tight')
plt.show()

print("‚úÖ Amenities impact plot saved!")
print(f"\nüíé Amenity Value Added:")
print(f"   Pool: +{pool_data[1] - pool_data[0]:,.0f} MAD")
print(f"   Garden: +{garden_data[1] - garden_data[0]:,.0f} MAD")
print(f"   Furnished: +{furnished_data[1] - furnished_data[0]:,.0f} MAD")
```

```python
# Cell 13: Summary Statistics
print("=" * 70)
print("üìã DATA EXPLORATION SUMMARY")
print("=" * 70)

print(f"\nüìä DATASET OVERVIEW:")
print(f"   Total Properties: {len(df):,}")
print(f"   Cities Covered: {df['City'].nunique()}")
print(f"   Property Types: {df['Property_Type'].nunique()}")
print(f"   Price Range: {df['Price_MAD'].min():,.0f} - {df['Price_MAD'].max():,.0f} MAD")

print(f"\nüèÜ TOP INSIGHTS:")
print(f"   1. Most expensive city: {city_stats.index[0]} (avg: {city_stats.iloc[0]['mean']:,.0f} MAD)")
print(f"   2. Most expensive property type: {type_stats.index[0]} (avg: {type_stats.iloc[0]['mean']:,.0f} MAD)")
print(f"   3. Size-Price correlation: {df['Size_m2'].corr(df['Price_MAD']):.2f} (strong positive)")
print(f"   4. Pool adds ~{pool_data[1] - pool_data[0]:,.0f} MAD to property value")
print(f"   5. {(df['Has_Pool'].sum() / len(df) * 100):.1f}% of properties have pools")

print(f"\n‚úÖ DATA QUALITY:")
print(f"   Missing Values: {df.isnull().sum().sum()} (0%)")
print(f"   Duplicate Rows: {df.duplicated().sum()}")
print(f"   Data Types: All correct")

print("\n" + "=" * 70)
print("‚úÖ DATA EXPLORATION COMPLETE!")
print("=" * 70)
```
